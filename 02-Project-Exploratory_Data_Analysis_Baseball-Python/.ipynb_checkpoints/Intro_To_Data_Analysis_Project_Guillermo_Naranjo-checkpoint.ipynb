## Exploration phase
* The first thing I would like to do is create single dataframe that contains the wins and salaries by team and year, grouped by team and year.

```python
#calculating aggregation of salaries by team by year
salaries_wo_players = salaries_clean.drop('playerID',1)
salaries_grp = salaries_wo_players.groupby(['yearID','teamID']).sum().reset_index()
teams_stats = salaries_grp.merge(teams_clean, on=['teamID','yearID'], how='inner')
teams_stats.head()
```

To prevent dealing with NetPresentValue to calculate the value of money today, I will just take the last 5 year for sake of comparison.

```python
teams_stats_recent = teams_stats[teams_stats['yearID'] >= 2011]
teams_stats_recent.head(10)
```

```python
teams_stats_recent.describe()
```

```python
def CalculateWinsByMillon(wins,salary):
    return float(wins / float(salary / 1000000.0))
```

```python
#Adding a new colum: wins by salary ratio in millions
teams_stats_recent.loc[:,('WinByMM')] = teams_stats_recent.apply(lambda row: CalculateWinsByMillon(row['W'], row['salary']), axis=1)
teams_stats_recent = teams_stats_recent.sort_values(by='WinByMM', ascending=False)
teams_stats_recent.head(10)
```

Now, lets remove outlier and see a graph without removing outliers.

```python
#Removing outliers, defining a function
def remove_outliers(dataset, field):
    Q1 = dataset[field].quantile(0.25)
    Q3 = dataset[field].quantile(0.75)    
    IQR = Q3 - Q1
    not_outliers = (dataset[field] > Q1 - 1.5 * IQR) & (dataset[field] < Q3 + 1.5 * IQR)
    return dataset[not_outliers]
```

```python
def isOutlier(dataset, field):
    Q1 = dataset[field].quantile(0.25)
    Q3 = dataset[field].quantile(0.75)    
    IQR = Q3 - Q1
    outliers = (dataset[field] < Q1 - 1.5 * IQR) | (dataset[field] > Q3 + 1.5 * IQR)
    return outliers
```

```python
teams_stats_no_outliers = remove_outliers(teams_stats_recent,'WinByMM')
```

```python
fig, ax = plt.subplots()
plt.hist(teams_stats_no_outliers['WinByMM'], bins=30, normed=False)
plt.xlabel('Wins by Million in salaries')
plt.ylabel('# of Teams')
plt.title('Team with the most wins by salary in Millions by season without outliers')
plt.grid(True)
plt.show()
```

### Now lets see the correlation between wins and salary

```python
fig, axis = plt.subplots()
# Grid lines, Xticks, Xlabel, Ylabel

axis.yaxis.grid(True)
axis.set_title('Salary in Millions vs Wins',fontsize=10)
axis.set_xlabel('Salary in $',fontsize=10)
axis.set_ylabel('Wins by Season',fontsize=10)

X = teams_stats_no_outliers['salary']
Y = teams_stats_no_outliers['W']

axis.scatter(X, Y)
plt.show()
```
